```python
#Min year is 92. I'm going to assume its 1992 and change it.
songs['Year'].replace(92,1992, inplace=True)
songs = songs[songs.Year.notna()] #must have a year
```

```python
#year should be an int, not float
songs["Year"] = songs['Year'].apply(lambda x: int(x))
```

```python
#half of danceability and valence is null. lesson says to drop when less than 50%
songs.dropna(thresh=len(songs.axes[1])/2,axis=0,inplace=True)
songs.drop(columns=['Danceability','Valence'],inplace=True)
```

```python
songs['Length (Duration)'].unique()
```

```python
#Length is object because of some commas for thousands numbers. should be int.
#remove commas, join it back.
for i in range(0,len(list(songs.index))):
    try: 
        if ',' in songs.iloc[i]['Length (Duration)']:
            j = songs.iloc[i]['Length (Duration)'].split(',')
            no_comma =''.join(j)
            songs.replace(songs.iloc[i]['Length (Duration)'],no_comma,inplace=True,limit=1)
    except:
        pass
```

```python
#Month is object and some months are spelled. should be int.
#for look to go through Month and replace the abbr with correct number
month_dict = {"Jan":1,"Feb":2,"Mar":3,"Apr":4,"May":5,"Jun":6,"Jul":7,"Aug":8,"Sep":9,"Oct":10,"Nov":11,"Dec":12}

for i in range(0,len(list(songs.index))):
    try:
        if songs.iloc[i]['Month'] in list(month_dict.keys()):
            songs.replace(songs.iloc[i]['Month'], str(month_dict[songs.iloc[i]['Month']]), inplace = True)
    except:
        pass #keep going

#Tom walked through a quicker way using %b with datetime. Much less code! Check zoom recording from 9/6.
```

```python
#cast them both to integers
songs["Month"] = songs['Month'].apply(lambda x: int(x))
```

```python
#songs['Length (Duration)'] = songs['Length (Duration)'].apply(lambda x: int(x))
#There are nulls in length (duration, so a lambda function isn't going to work. Need to cast this as an int in a different way.
#pandas has to_numeric. I think it skips nulls and lets them be!
songs['Length (Duration)'] = pd.to_numeric(songs['Length (Duration)'],downcast='integer')
```

```python
# Calculate descriptive statistics for each column
# Let's define an outlier as a value at least 3 standard deviations away from the mean. 
# Which columns have outliers? What are those values?

#mean +/- 3 standard deviations
#for loop going through each column. if 3 std more or les than mean.
#inspo: https://stackoverflow.com/questions/23199796/detect-and-exclude-outliers-in-a-pandas-dataframe?noredirect=1&lq=1
def outlier(input, mean, std):
    if ((mean + 3*std) < input) or ((mean - 3*std)> input):
        return [True, input]
    else:
        return [False, input]


 #go through each column in the list, if in that column, the value is an outlier, append it to the list. if not, move   
for x in list(songs.columns):
    try:
        outlier_list = []
        print(x)
        for y in songs[x]:
            if outlier(y,songs[x].mean(),songs[x].std())[0]:
                outlier_list.append(outlier(y,songs[x].mean(),songs[x].std())[1])
        print(outlier_list)
    except:
        continue
```

```python
#With the remaining columns, use scikit-learn to impute missing values.
# For continuous features, fill in the mean. 
# For categorical features, fill in the mode. mode is most frequent strategy

#inspo: https://scikit-learn.org/stable/modules/impute.html
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

mean_imp = SimpleImputer(missing_values=np.nan, strategy='mean')

mode_imp = SimpleImputer(strategy='most_frequent')

for x in list(songs.columns)[1:]:
    if type(songs[x][0])==type(float) or type(songs[x][0])==type(int):
        songs[x] = mean_imp.fit_transform(songs[x].values.reshape(-1,1))
    else:
        songs[x] = mode_imp.fit_transform(songs[x].values.reshape(-1,1))

#Reminder: This might not be accurate for the data but it will help the machine learning.
#This will input the mean length into a song w/ a null length. That's not the length of the song but it will be a point for ML to work with.
```

```python
# Combine the date-related columns into one column with the Pandas to_datetime() method, then use that column to create a numeric Age column (in years). 
# Calculate Age based on today's date; it doesn't have to be a whole number. 
# Once you've created the Age column, remove the other date-related columns, including the one you created with Pandas.
```

```python
import datetime
datetime_strs = []
for indx, row in songs.iterrows():
    datetime_strs.append(datetime.date(row['Year'],row['Month'],1))

songs['Combined Dates'] = datetime_strs
```

```python
#https://www.geeksforgeeks.org/python-datetime-toordinal-method-with-example/
#toordinal is number of days since year 1! 
#Take num of days since today, minus num of days since the DATE. Gives x number of days since the song has been out. divide by 365 to see how many years.
today = datetime.datetime.today().toordinal()

ages =[]
for dates in songs['Combined Dates'].values:
    ages.append((today-dates.toordinal())/365)

songs['Age'] = ages
```
